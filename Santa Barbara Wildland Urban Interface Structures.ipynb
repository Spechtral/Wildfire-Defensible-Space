```python
import arcgis
from arcgis.gis import GIS
# Create a GIS object, as an anonymous user for this example
gis = GIS()
# Create a map widget
map1 = gis.map('Santa Barbara, CA', 12.75) # Passing a place name to the constructor
                        # will initialize the extent of the map.
map1
```

```python
# add the layer depicting structure footprints and their home iginition zones (this is a clip of a custom layer I made in ArcGIS Pro)
world_countries_item = gis.content.get('6c111a9027454983adb361b0be0fcfce')
world_countries_layer = world_countries_item.layers[0]
world_countries_layer

map1.add_layer(world_countries_layer)
```
